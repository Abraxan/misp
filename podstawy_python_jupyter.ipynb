```python
#### challenge 0.2

zajecia = [["wf"],
           ["informatyka", "nauka czytania"],
           ["relax", "jedzenie"],
           ["mispy","mispy w domu"],
           ["piateczek", "praca domowa z mispow", "Ilovemispy", "mispy w czasie snu"]
          ]

dnityg= ["Pon","Wto","Sro","Czwar","Pio"]

# zdefiniuj funkcje ktora poda ile jest zajec danego dnia
# lista_zajec_z_dnia -> ile ich jest
def ilosc_zajec(dany_dzien):
    return len(dany_dzien)

# po
# posortuj dni po najbardziej zajetych
  #łączenie listy w lstę z 5-cioma krotkami
dany_dzien_zajecia = list(zip(dnityg, zajecia))
  #posortowanie zajęc od najbardziej zajętych do najmniej, wykorzystując reverse=True to odwrócenia kolejności listy jaką stworzyła funkcja sorted
dany_dzien_zajecia = sorted(dany_dzien_zajecia, key= lambda sort:ilosc_zajec(sort[1]), reverse=True)
  #wyświetlenie wyniku
print(dany_dzien_zajecia)

print("-----------------")

#### challenge 0.3

# napisz program ktory zdajdzie dzien z najwieksza iloscia mispow
def mispy_na_dzien(dany_dzien):
   #deklaracja licznika
  i=0
    #sprawdzenie czy w nazwie przedmiotu jest "misp" dla każdego dnia po kolei
  for zajecie in dany_dzien:
    if zajecie.find("misp")==-1:
      continue
    i=i+1
  return i

  #sortowanie za pomocą klucza i funkcji sorted
dany_dzien_zajecia = sorted(dany_dzien_zajecia, key=lambda sort:mispy_na_dzien(sort[1]))
  #wyświetlanie wyniku biorąc pod uwagę, że w powyższej linijce funkcja sorted zwróciła kolejność elementów w liście od tych, które mają najmniej mispów
print("Najwięcej mispów jest w: {}".format(dany_dzien_zajecia[-1][0]))
```
