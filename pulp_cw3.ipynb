```
!pip install pulp
# Import PuLP modeler functions
from pulp import *
import pandas as pd

  #tworzenie zmiennej zawierającej dane problemu
problem = LpProblem("Mostek",LpMinimize)

  #zmienne decyzyjne prądów
I1 = LpVariable("Prad_1",0)
I2 = LpVariable("Prad_2",0)
I3 = LpVariable("Prad_3",0)
I4 = LpVariable("Prad_4",0)
I5 = LpVariable("Prad_5",0)

  #ograniczenia
problem += 6*I1 + 10*I2 + 4*I3 + 7*I4 + 3*I5, "Moc"
problem += 3 <= I1, "Prad_1_min"
problem += I1 <= 5, "Prad_1_max"
problem += 1 <= I2, "Prad_2_min"
problem += I2 <= 3, "Prad_2_max"
problem += 1 <= I3, "Prad_3_min"
problem += I3 <= 3, "Prad_3_max"
problem += 1 <= I4, "Prad_4_min"
problem += I4 <= 3, "Prad_4_max"
problem += 3 <= I5, "Prad_5_min"
problem += I5 <= 5, "Prad_5_max"

  #zapisanie danych do pliku .lp
problem.writeLP("MostekModel.lp")

  #rozwiązanie problemu w pulp'ie
problem.solve()

  #wyświetlenie statusu rozwiązania
print ("Status:", LpStatus[problem.status])

  #wyświetlenie zmiennych z ich optymalną wartością
for v in problem.variables():
    print (v.name, "=", v.varValue)

  #wyświetlenie otrzymanych wartości rezystancji
print("Wartosc pierwszej rezystancji wynosi", 6/I1.varValue,"k Ohm")
print("Wartosc drugiej rezystancji wynosi", 10/I2.varValue,"k Ohm")
print("Wartosc trzeciej rezystancji wynosi", 4/I3.varValue,"k Ohm")
print("wartosc czwartej rezystancji wynosi", 7/I4.varValue,"k Ohm")
print("wartosc piatej rezystancji wynosi", 3/I5.varValue,"k Ohm")
```
