```
  #sprawdznie jaki procent kobiet i meżczyzn przeżyło na podstawie danych treningowych
panie = trening.loc[trening.Sex == 'female']["Survived"]
wsk_panie = sum(panie)/len(panie)
panowie = trening.loc[trening.Sex == 'male']["Survived"]
wsk_panowie = sum(panowie)/len(panowie)

  #wyświetlenie danych
print("Kobiety, które przeżyły: {} %".format(wsk_panie*100))
print("Mężczyźni, którzy przeżyli: {} %".format(wsk_panowie*100))
```

```
trening["Age"].fillna(trening.Age.mean(), inplace=True)
trening.isnull().sum()
```

```
 #wyznaczenie celu i interesujących danych
features = trening[["Pclass", "Sex", "Age",'Parch', 'SibSp']]
features['Sex']=features['Sex'].replace('male', 0)
features['Sex'] =features['Sex'].replace('female', 1)
features.head(5) 
```

```
  #importowanie bibliotech do uczenia maszynowego
    #róźne modele
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
    #narzędzia - wznaczanie błędu bezwzględnego i dzielenia danych
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

  #tworzenie modeli
decision_tree_model= DecisionTreeRegressor(random_state=1)
tree_classifier_model= DecisionTreeClassifier(random_state=1)
neighbor_model= KNeighborsClassifier(5)
svc_model= SVC(random_state=1)
forest_model= RandomForestClassifier(random_state=1)

  #ustawienie celu
y = trening[["Survived"]]
  #dzielenie danych
train_X, val_x, train_y, val_y = train_test_split(features, y, random_state=1)

  #dopasowanie modeli
decision_tree_model.fit(train_X, train_y)
tree_classifier_model.fit(train_X, train_y)
neighbor_model.fit(train_X, train_y)
svc_model.fit(train_X, train_y)
forest_model.fit(train_X, train_y)

  #sprawdzanie poprawności modeli
decision_tree_val_mea = mean_absolute_error(val_y, decision_tree_model.predict(val_x))
tree_classifier_val_mea = mean_absolute_error(val_y, tree_classifier_model.predict(val_x))
svc_val_mea = mean_absolute_error(val_y, svc_model.predict(val_x))
neighbor_val_mea = mean_absolute_error(val_y, neighbor_model.predict(val_x))
forest_val_mea = mean_absolute_error(val_y, forest_model.predict(val_x))

print("Błąd bezwzględny:")
print(" *DecisionTreeRegressor: ",decision_tree_val_mea)
print(" *DecisionTreeClassifier: ",tree_classifier_val_mea)
print(" *SVC: ",svc_val_mea)
print(" *KNeighborsClassifier: ",neighbor_val_mea)
print(" *RandomForestClassifier: ", forest_val_mea)
```

Najmniejszy błąd jest dla RandomForestClassifier, które zostanie wykorzystane dla pliku testowego
