```
from pulp import *
prob = LpProblem("The_Whiskas_Problem",LpMinimize)
LpVariable("example", None, 100)

# The 2 variables Beef and Chicken are created with a lower limit of zero + ryba
x1=LpVariable("ChickenPercent",0,None,LpInteger)
x2=LpVariable("BeefPercent",0)
x3=LpVariable("FishPercent",0)

# The objective function is added to 'prob' first
prob += 0.013*x1 + 0.008*x2 + 0.01*x3, "Total Cost of Ingredients per can"

# The five constraints are entered
prob += x1 + x2 + x3 == 100, "PercentagesSum"
prob += 0.100*x1 + 0.200*x2 + 0.15*x3 >= 8.0, "ProteinRequirement"
prob += 0.080*x1 + 0.100*x2 + 0.075*x3 >= 6.0, "FatRequirement"
prob += 0.001*x1 + 0.005*x2 + 0.008*x3 <= 2.0, "FibreRequirement"
prob += 0.002*x1 + 0.005*x2 + 0.003*x3 <= 0.4, "SaltRequirement"

# The problem data is written to an .lp file
prob.writeLP("WhiskasModel.lp")

# The problem is solved using PuLP's choice of Solver
prob.solve()

# The status of the solution is printed to the screen
print("Status:", LpStatus[prob.status])

# Each of the variables is printed with it's resolved optimum value
for v in prob.variables():
    print(v.name, "=", v.varValue)

# The optimised objective function value is printed to the screen
print("Total Cost of Ingredients per can = ", value(prob.objective))
```

```
from pulp import *
# lista składników
Ingredients = ['Chicken', 'Beef', 'Mutton', 'Rice', 'Wheat bran', 'Gel','Sardines','Carrot']

#koszty poszczegołnych składników
costs = {'Chicken': 0.013, 
         'Beef': 0.008, 
         'Mutton': 0.010, 
         'Rice': 0.002, 
         'Wheat bran': 0.005, 
         'Gel': 0.001,
         'Sardines': 0.009,
         'Carrot':0.003}

#procent protein w danym składniku
proteinPercent = {'Chicken': 0.100, 
                  'Beef': 0.200, 
                  'Mutton': 0.150, 
                  'Rice': 0.000, 
                  'Wheat bran': 0.040, 
                  'Gel': 0.000,
                  'Sardines':0.120,
                  'Carrot':0.001}

#procent tłuszczu w danym składniku
fatPercent = {'Chicken': 0.080, 
              'Beef': 0.100, 
              'Mutton': 0.110, 
              'Rice': 0.010, 
              'Wheat bran': 0.010, 
              'Gel': 0.000,
              'Sardines': 0.090,
              'Carrot':0.000}

#procent błonnika w danym składniku
fibrePercent = {'Chicken': 0.001, 
                'Beef': 0.005, 
                'Mutton': 0.003, 
                'Rice': 0.100, 
                'Wheat bran': 0.150, 
                'Gel': 0.000,
                'Sardines': 0.004,
                'Carrot':0.200}

#procent soli w danym składniku
saltPercent = {'Chicken': 0.002, 
               'Beef': 0.005, 
               'Mutton': 0.007, 
               'Rice': 0.002, 
               'Wheat bran': 0.008, 
               'Gel': 0.000,
               'Sardines': 0.001,
               'Carrot': 0.002}

#procent witaminy B w danym składniku
Bvitamin = {'Chicken': 0.3 * 0.01, 
            'Beef': 2.6 * 0.01, 
            'Mutton': 2.6 * 0.01, 
            'Rice': 0 * 0.01, 
            'Wheat bran': 0 * 0.01, 
            'Gel': 0 * 0.01,
            'Sardines': 8.9 * 0.01,
            'Carrot': 3.1 * 0.01}

#zmienna prob zawiera dane problemu
prob = LpProblem("The Whiskas Problem", LpMinimize)

#słownik zawierający odniesienie do zmiennych
ingredient_vars = LpVariable.dicts("Ingr",Ingredients,0)

#występujące ograniczenia
prob += lpSum([costs[i]*ingredient_vars[i] for i in Ingredients]), "Total Cost of Ingredients per can"
prob += lpSum([ingredient_vars[i] for i in Ingredients]) == 100, "PercentagesSum"
prob += lpSum([proteinPercent[i] * ingredient_vars[i] for i in Ingredients]) >= 8.0, "ProteinRequirement"
prob += lpSum([fatPercent[i] * ingredient_vars[i] for i in Ingredients]) >= 7.0, "FatRequirement"
prob += lpSum([fibrePercent[i] * ingredient_vars[i] for i in Ingredients]) >= 2.0, "FibreRequirement"
prob += lpSum([saltPercent[i] * ingredient_vars[i] for i in Ingredients]) <= 1.0, "SaltRequirement"
prob += lpSum([Bvitamin[i] * ingredient_vars[i] for i in Ingredients]) >= 1.8, "BVitaminRequirement" 

#dane zapisane jako plik .lp
prob.writeLP("WhiskasModel2.lp")

#rozwiązanie problemu w pulp'ie
prob.solve()

#wyświetlenie statusu rozwiązania
print("Status:", LpStatus[prob.status])

#wyświetlenie każdego składnika z optymalną wartością
for v in prob.variables():
    print(v.name, "=", v.varValue)

#wyświetlenie optymalnego kosztu składników w przeliczeniu na jedną puszkę karmy
print("Total Cost of Ingredients per can = ", value(prob.objective))
```
